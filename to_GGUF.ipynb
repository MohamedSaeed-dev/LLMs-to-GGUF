```python
from huggingface_hub import HfApi

api = HfApi()

model_id = "MohamedSaeed-dev/gemma-7b-gguf"  # Specify the correct model ID
repo_type = "model"  # Specify the correct repo type

# Create repository
api.create_repo(repo_id=model_id, exist_ok=True)

# Upload file to the repository
api.upload_file(
    path_or_fileobj="gemma-7b-v1.5.gguf",
    path_in_repo="gemma-7b-v1.5.gguf",
    repo_id=model_id,
    repo_type=repo_type
)
```
